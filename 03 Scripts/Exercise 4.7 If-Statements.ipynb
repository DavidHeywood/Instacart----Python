# If-Statements

```python
# Defining the function price_label
# This function categorizes items into price ranges: low, mid, and high.

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Running price_label on df.
# df['price_range'] stores the new labels in a new column in df called 'price_range'
# axis=1 runs the function on all rows; axis=0 would run it on all columns.

df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

## Step 2: re-thinking the 'busiest_day' column

Using the frequency table for 'orders_day_of_week' above, we see that the busiest days are Saturday and Sunday (days 0 and 1), and the slowest days are day Tuesday and Wednesday (days 3 and 4).

```python
# Creating an empty list into which the values of our for-loop will be stored
result_2 = []
```

```python
# Filling 'result_2' with 'Busiest', 'Least busy', or 'Regularly busy'
for x in df_ords_prods_merged["orders_day_of_week"]:
  if x == 0 or x == 1:
    result_2.append("Busiest")
  elif x == 3 or x == 4:
    result_2.append("Least busy")
  else:
    result_2.append("Regularly busy")
```

```python
result_2
```

```python
# Combining the 'result_2' list with df_ords_prods_merged
df_ords_prods_merged['busiest_days'] = result_2
```

## Step 3: checking the newly-created 'busiest_days' column

```python
# Frequency of 'busiest_days'
df_ords_prods_merged['busiest_days'].value_counts(dropna = False)
```

The busiest days were days 0 and 1, with frequencies of 6204182 and 5660230, respectively.  The sum of those frequencies matches the frequency of rows labeled "Busiest," so the result of the 'busiest_days' column is correct.  The other two categories also add up.  So it seems that all went according to plan!

## Step 4: creating a 'busiest_period_of_day' column

```python
# Getting a frequency of 'order_hour_of_day' column
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

Since three categories (most orders, average orders, and fewest orders) are desired, I will split the 24 hours of the day into 8 hour blocks.  That is, 8 hours with the highest volume of sales will be labeled 'Most orders,' the 8 hours with the lowest volume will be labeled 'Fewest orders,' and the other 8 hours will be labeled 'Average orders.'

Note that the highest volume occurs from hours 9-16.  The lowest volume occurs from hours 23-6.  The middle volume occurs from hours 7-8 and 17-22.

```python
# Creating an empty list into which the values of our for-loop will be stored
result_3 = []
```

```python
# Filling 'result_3' with 'Most orders', 'Fewest orders', or 'Average orders'

for x in df_ords_prods_merged["order_hour_of_day"]:
  if x in range(9,17):                  #range will stop at 16, as desired
    result_3.append("Most orders")
  elif (x in range(0,7)) or x == 23:
    result_3.append("Fewest orders")
  else:
    result_3.append("Average orders")
```

```python
result_3
```

```python
# Combining the 'result_3' list with df_ords_prods_merged
df_ords_prods_merged['busiest_period_of_day'] = result_3
```

## Step 5: printing the frequency of 'busiest_period_of_day'

```python
# Frequency of 'busiest_period_of_day'
df_ords_prods_merged['busiest_period_of_day'].value_counts(dropna = False)
```

Again, the numbers add up here.
