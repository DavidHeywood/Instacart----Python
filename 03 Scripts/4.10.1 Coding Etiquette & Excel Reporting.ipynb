# Crosstabs

```python
# Creates a crosstab between 'days_since_prior_order' and 'order_number' to check if all missing values from 'days_since_prior_order' have an 'order_number' of 1.

crosstab = pd.crosstab(df['days_since_prior_order'], df['order_number'], dropna = False)
```

```python
# To view the crosstab we just created, we first copy it to the clipboard.

crosstab.to_clipboard()

# Then we paste it into an Excel workbook.
```
